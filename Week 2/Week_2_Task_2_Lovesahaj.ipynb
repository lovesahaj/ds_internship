```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```
df = pd.read_csv('EmployeeAttrition.csv')
df.head()
```

```
df.columns
```

```
df["Attrition"] = df["Attrition"].apply(lambda x : 1 if x == "Yes" else 0)
```

```
travel_dict = {'Travel_Rarely' : 1, 'Travel_Frequently' : 2, 'Non-Travel' : 0}
```

```
df['BusinessTravel'] = df['BusinessTravel'].apply(lambda x : travel_dict[x])
```

```
df.hist(figsize=(20, 20))
plt.show()
```
